```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("F_phoneme27_AM89_MnasNet_times0.csv")
```

```python
data
```

```python
data["error"] = data[" prediction"] - data[" label"]
```

```python
data["mae"] = np.abs(data["error"])
```

```python
data["se"] = np.square(data["error"])
```

```python
sns.scatterplot(data=data, x=" length", y="mae", s=5)
```
